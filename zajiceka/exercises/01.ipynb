```python
from rdkit import Chem
from rdkit import IPythonConsole
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
import numpy as np
```

```python
m = Chem.MolFromSmiles('O=C(C)Oc1ccccc1C(=O)O')
m
```

```python
m1 = Chem.MolFromSmiles('CC(=O)OC1=CC=CC=C1C(=O)[O-]')
m1
```

```python
print(Chem.MolToSmiles(m))
print(Chem.MolToSmiles(m1))
```

```python
Chem.MolToInchi(m1)
```

```python
print(Chem.MolToInchiKey(m))
print(Chem.MolToInchiKey(m1))
```

```python
print(Chem.MolToMolBlock(m))
```

```python
Chem.GetSSSR(m)
```

```python
m3 = Chem.MolFromSmiles('C12C3C4C1C5C2C3C45')
m3 = Chem.AddHs(m3)
AllChem.EmbedMolecule(m3)
Draw.MolToMPL(m3)
```

```python
Chem.GetSSSR(m3)
```

```python
mols =  Chem.SDMolSupplier('../../data/drugbank.sdf')
drugbank = [mol for mol in mols if mol]
```

```python
rings = np.array([mol.GetRingInfo().NumRings() for mol in drugbank])
```

```python
rings = np.array([Chem.GetSSSR(mol) for mol in drugbank])
```

```python
print(rings.sum())
print(rings.mean())
```
